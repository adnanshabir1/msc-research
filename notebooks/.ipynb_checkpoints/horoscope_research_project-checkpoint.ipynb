##### Convert Date to Proper Format

```python
# Convert date to proper datetime format
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

# Check result
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

##### Clean the Horoscope Text

```python
import re  # regular expressions (used for text cleaning)

# Define a cleaning function
def clean_text(text):
    text = text.lower()                    # convert all text to lowercase
    text = re.sub(r'[^\w\s]', '', text)    # remove punctuation
    return text

# Apply the function to create a new column
df['clean_text'] = df['text'].apply(clean_text)

# View results
df[['text', 'clean_text']].head()
```

### Baseline Model - VADER 

#### Apply Sentiment Analysis (VADER)

```python
# Apply sentiment analysis to clean_text

# Create a new column for sentiment scores
df['sentiment'] = df['clean_text'].apply(lambda x: sia.polarity_scores(x)['compound'])

# View results
df[['clean_text', 'sentiment']].head()
```

##### Convert Sentiment Into Categories

```python
# Function to classify sentiment
def classify_sentiment(score):
    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"

# Apply classification
df['sentiment_label'] = df['sentiment'].apply(classify_sentiment)

# View results
df[['sentiment', 'sentiment_label']].head()
```

##### Analyze Sentiment Distribution

```python
# Count how many of each sentiment type
sentiment_counts = df['sentiment_label'].value_counts()

print(sentiment_counts)
```

```python
# Convert counts to percentage
sentiment_percent = df['sentiment_label'].value_counts(normalize=True) * 100

print(sentiment_percent)
```

Horoscope texts are heavily biased toward positive language

#### 5. Add Mercury Retrograde Labels

```python
# Mercury retrograde periods (2020 + 2021)
retrograde_periods = [
    # 2020
    ('2020-06-18', '2020-07-12'),
    ('2020-10-14', '2020-11-03'),
    
    # 2021
    ('2021-01-30', '2021-02-20'),
    ('2021-05-29', '2021-06-22')
]
```

```python
# Function to check if a date falls in retrograde
def is_retrograde(date):
    for start, end in retrograde_periods:
        if pd.to_datetime(start) <= date <= pd.to_datetime(end):
            return 1
    return 0
```

```python
# Apply the function
df['retrograde'] = df['date'].apply(is_retrograde)
```

```python
# Check result
df[['date', 'retrograde']].head(10)
```

##### Compare Sentiment (Retrograde vs Normal)

```python
# Calculate average sentiment for each group
sentiment_comparison = df.groupby('retrograde')['sentiment'].mean()

print(sentiment_comparison)
```

```python
# Plot comparison
sns.boxplot(x='retrograde', y='sentiment', data=df)

plt.title("Sentiment Distribution: Retrograde vs Normal")
plt.xlabel("Retrograde (0 = No, 1 = Yes)")
plt.ylabel("Sentiment Score")

plt.show()
```

Note: horoscope sentiment becomes slightly more negative during Mercury retrograde but difference is very small

```python
# Import pipeline
from transformers import pipeline

# Load sentiment analysis model (this uses BERT under the hood)
bert_classifier = pipeline("sentiment-analysis")

print("BERT model loaded successfully")
```

```python
text = "Things may go wrong today"

bert_classifier(text)
```

```python
bert_classifier("I feel great today!")
```

```python
# Take a random sample of 500 rows
df_sample = df.sample(500, random_state=42).reset_index(drop=True)

# View sample
df_sample.head()
```

```python
# Apply BERT to the sample dataset
bert_results = df_sample['text'].apply(lambda x: bert_classifier(x)[0])

# Convert results into separate columns
df_sample['bert_label'] = bert_results.apply(lambda x: x['label'])
df_sample['bert_score'] = bert_results.apply(lambda x: x['score'])

# View results
df_sample[['text', 'bert_label', 'bert_score']].head()
```

##### Compare Sentiment During Retrograde (Using BERT)

```python
# Convert BERT labels to numeric values
def convert_bert_label(label):
    if label == "POSITIVE":
        return 1
    else:
        return -1

df_sample['bert_numeric'] = df_sample['bert_label'].apply(convert_bert_label)

# Compare sentiment during retrograde vs normal
bert_comparison = df_sample.groupby('retrograde')['bert_numeric'].mean()

print(bert_comparison)
```

```python
from scipy.stats import ttest_ind

# Separate groups
bert_retro = df_sample[df_sample['retrograde'] == 1]['bert_numeric']
bert_normal = df_sample[df_sample['retrograde'] == 0]['bert_numeric']

# Perform T-test
t_stat, p_value = ttest_ind(bert_retro, bert_normal)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```
